# Healthcare Appointments Analysis
This notebook cleans the dataset, performs exploratory data analysis, and extracts business insights to optimize hospital operations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
plt.style.use('ggplot')
```

## 1. Load Data
We'll load the Appointment data and perform an initial review.

```python
df = pd.read_excel('Healthcare Appointments.xlsx')
print(f'Total Records: {len(df)}\n')
display(df.head())
```

## 2. Data Cleaning
We want to ensure all our records have logical values. We handle missing strings and ensure dates are parsed correctly.

```python
df['Appointment Reason'] = df['Appointment Reason'].fillna('Not Specified')
df['Reason Category'] = df['Reason Category'].fillna('Not Specified')

# Ensure time/date columns format correctly if needed
df['Appointment Date'] = pd.to_datetime(df['Appointment Date'], errors='coerce')
display(df.info())
```

## 3. Analysis & Insights
### A. Appointment Status (No-Show vs Completed)
Identifying the drop-off rate is crucial for determining how much revenue or time is lost to No-Shows.

```python
status_counts = df['Appointment Status'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=status_counts.index, y=status_counts.values, palette='Blues_r')
plt.title('Distribution of Appointment Statuses')
plt.ylabel('Count')
plt.show()

print('\nBusiness Insight: High no-show rates can be addressed by implementing automated SMS/email reminders 24-48 hours prior to appointments.')
```

### B. Busiest Providers
Monitoring the workload of providers avoids burnout and improves patient care.

```python
provider_counts = df['Provider Name'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(y=provider_counts.index, x=provider_counts.values, palette='viridis')
plt.title('Top 10 Busiest Providers')
plt.xlabel('Number of Appointments')
plt.show()

print('\nBusiness Insight: Providers handling the majority of appointments might experience burnout. Workload balancing strategies could help distribute these cases more evenly across available medical staff.')
```

### C. Top Reason Categories
What are the main reasons patients are visiting?

```python
reason_trends = df['Reason Category'].value_counts().head(8)

plt.figure(figsize=(10,5))
sns.barplot(x=reason_trends.values, y=reason_trends.index, palette='flare')
plt.title('Most Common Appointment Reason Categories')
plt.xlabel('Number of Appointments')
plt.show()

print('\nBusiness Insight: Identifying the most common categories helps the practice properly stock medical supplies and allows marketing teams to focus on preventive care campaigns for the most frequent issues.')
```
