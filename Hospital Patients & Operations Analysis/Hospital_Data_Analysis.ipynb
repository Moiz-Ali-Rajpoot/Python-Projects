# Hospital Data Analysis
This notebook cleans the Hospital dataset, explores factors related to hospital stays, admissions, and generates insights for better hospital management.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set_theme(style='darkgrid')
```

## 1. Load Data
We load the Hospital Data CSV which contains detailed information on admissions, stays, and severity of illnesses.

```python
df = pd.read_csv('Hospital Data.csv')
print(f'Total Records: {len(df)}\n')
display(df.head())
```

## 2. Data Cleaning
We want to check for and handle missing values appropriately, particularly in the 'Health Conditions' column which has a lot of nulls.

```python
print('Missing values before cleaning:\n', df.isnull().sum())

df['Health Conditions'] = df['Health Conditions'].fillna('None/Undisclosed')
print('\nMissing values after cleaning:\n', df.isnull().sum())
```

## 3. Exploratory Data Analysis
### A. Department Workloads
Understanding patient distribution across departments helps with resource allocation.

```python
dept_counts = df['Department'].value_counts()

plt.figure(figsize=(10,6))
sns.barplot(x=dept_counts.values, y=dept_counts.index, palette='mako')
plt.title('Patient Distribution by Department')
plt.xlabel('Number of Patients')
plt.ylabel('Department')
plt.show()

print('\nBusiness Insight: The department with the highest patient volume should be allocated the most beds and staff to minimize bottlenecks and reduce wait times.')
```

### B. Severity of Illness vs. Stay Duration
Does a more severe illness translate to a systematically longer hospital stay?

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Severity of Illness', y='Stay (in days)', order=['Minor', 'Moderate', 'Extreme'], palette='Set2')
plt.title('Length of Stay by Illness Severity')
plt.show()

print('\nBusiness Insight: Patients with Extreme severity tend to have larger variances and longer stays. Financial planning and bed availability predicting models should weight extreme cases heavily to avoid resource exhaustion.')
```

### C. Age Distribution in Hospital Admissions
Which age brackets make up the bulk of admissions?

```python
age_dist = df['Patient Age'].value_counts().sort_index()

plt.figure(figsize=(10,6))
sns.barplot(x=age_dist.index, y=age_dist.values, palette='magma')
plt.title('Demographics: Age Bracket of Admitted Patients')
plt.xlabel('Age Bracket')
plt.ylabel('Count')
plt.show()

print('\nBusiness Insight: Knowing the primary age demographic helps tailor the hospital\'s amenities, meal plans, and the types of specialized physicians kept on call.')
```
